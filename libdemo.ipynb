```python
"""
Let's start with the first piece

"""

from sockproc.serversock import MasterProcess
master = MasterProcess('127.0.0.1', 8000)



"""
We've just created an instance of the object described in 1.  Our object
accomplishes those bullet points, but how?

1) 
    a) shared resource
        Prior to being able to simply import an object from this package like shown,
        we need to install our package.
    
            "cd /path/to/this/package"
            "python setup.py install"
    
        MasterProcess checks for and creates a directory called "/etc/ipc" for us to utilize.
        That directory will contain two things:
            - A text file identifying our host and port for our MasterProcess 
              socket that we'll be using to communicate with child processes
              This is where our processes will look for the information they need
              to spawn their own sockets to communicate with the master through.
              
            - A text file that will contain the process ids of all our
              child processes.  The second part is redundant because there is also an interable
              attribute of the MasterProcess object that contains those same process ids.
    

       
       
    b) master socket to spawn client sockets and receive responses
    
        The two arguments to MasterProcess are host and port.  The host needs to be '127.0.0.1',
        but port can be any port.  This is the host and port that our socket object will be bound
        to.  This socket object is the "server socket".  It will spawn client sockets for our processes
        to communicate with.  I would refrain from using privileged ports < 1024, though.
        
        After we've established a connection with a process via a socket, information can freely
        flow back and forth between those process specific sockets.  I can't speak to their performance
        in relation to other IPC methods, though.
        
        
    c) smooth process termination to give control of the shell back to the process which spawned
       all the processes
       
           /etc/ipc/pids.txt
           
               - this file will be full of our process ids, and will be parsed to kill off those
                 poor processes
                 
"""
```

```python
# a brief digression on sockets
import socket

host = '127.0.0.1'
port = 5555

mastersock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)


print mastersock.getsockname()

# bind our mastersock to the above
mastersock.bind( (host, port) )


print mastersock.getsockname()
# set the number of connections
mastersock.listen(2)


# Example of what gets created in our Process when we spawn one with this package

first_process = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
first_process.connect( (host, port) )


"""
When first_process called 'connect()' with the mastersock's host and port,
mastersock received a connection request.  When we accept that request our
mastersock or server socket returns to us a client socket, which is just a socket
whose sole purpose is to communicate with the socket that requested a connection.

Sockets are bidirectional, and this specific client socket will be communicating
only with the socket that made the request for it to exist.  

addr is just the host and port of the socket requestion a connection
"""

(client, addr) = mastersock.accept()

# let's send some data between the sockets
first_process.send('sockets are cool')

client.recv(14)


client.send('200')

first_process.recv(3)
```
